```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('C:/Users/user/Desktop/StudentsPerformance.csv')
data
```

# 1.How many males and females participated in the test.

```python
Male=data[data['gender']=="male"].count()
m= Male['gender']
Female=data[data['gender']=="female"].count()
f= Female['gender']
print("No:of Males: ", m)
print("No:of Females: ", f)
```

# 2.What do you think about the students parental level of education.

```python
plt.figure(figsize=(5,5))
val=data.groupby('parental level of education')['parental level of education'].count().sort_values()
plt.plot(val,color='blue',marker='x')  
plt.xticks(rotation=90)
plt.show()
```
